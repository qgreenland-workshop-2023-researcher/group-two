**MEaSUREs Greenland Annual Ice Sheet Velocity Mosaics**

This data is specific to Dec 2020 - Nov 2021 (Inclusive)

 Here we load the dataset into Python as an XArray object and browse the metadata:

```python
import numpy as np 
from pathlib import Path
import xarray as xra


# load velocity magnitude (vv) from folder
datafile_path = Path('/home/jovyan/shared-public/QGreenland-Researcher-Workshop/group-two/velocity2021/GL_vel_mosaic_Annual_01Dec20_30Nov21_vv_v04.0.tif')
ds = xra.open_dataset(datafile_path)
ds
```

#### Here we get a list of all the object functions available for the dataset

```python
# Show the available "public" methods and attributes provided by `rioxarray`:
print([i for i in dir(ds.rio) if not i.startswith('_')])
```

```python
print(f'CRS: {ds.rio.crs}')
print(f'Units: {ds.rio.crs.linear_units}')
print(f'Bounds: {ds.rio.bounds()}')
print(f'Height: {ds.rio.height}')
print(f'Width: {ds.rio.width}')
print(f'Grid pixel resolution: {ds.rio.resolution()}')
```
